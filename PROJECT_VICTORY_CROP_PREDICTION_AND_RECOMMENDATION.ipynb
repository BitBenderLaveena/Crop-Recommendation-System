```python
ms=MinMaxScaler()                       #All values in specific range


ms.fit(X_train)
X_train=ms.transform(X_train)
X_test=ms.transform(X_test)
```

#Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()                       #All values in specific range


sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

#Training Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

```

```python
#Creating instanaces of all models

models={'Logistic Regression':LogisticRegression(),         #Name-Model
        'Naive Bayes':GaussianNB(),
        'Support Vector Machine':SVC(),
        'K-Nearest Neighnors':KNeighborsClassifier(),
        'Decision Tree':DecisionTreeClassifier(),
        'Random Forest':RandomForestClassifier(),
        'Bagging':BaggingClassifier(),
        'AdaBoost':AdaBoostClassifier(),
        'Gradient Boosting':GradientBoostingClassifier(),
        'Extra Trees:':ExtraTreeClassifier()
        }
```

```python
for name, md in models.items():
  md.fit(X_train,y_train)
  ypred=md.predict(X_test)

  print(f"{name} with accuracy:{accuracy_score(y_test,ypred)}")
```

Random forest giving among max accuracy.

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
ypred=rfc.predict(X_test)
accuracy_score(y_test,ypred)
```

```python
def recommendation(N,P,K,crop_year,area,temperature,wind_speed,precipitation,humidity,production,Yield):
  features=np.array([[N,P,K,crop_year,area,temperature,wind_speed,precipitation,humidity,production,Yield ]])
  transformed_feature=ms.fit_transform(features)
  transformed_features=sc.fit_transform(transformed_feature)
  prediction=rfc.predict(transformed_features).reshape(1,-1)

  return prediction[0]
```
